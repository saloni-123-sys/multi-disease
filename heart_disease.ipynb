Splittig the features and target

```python
x = heart_disease_df.drop(columns='target',axis=1)
y = heart_disease_df['target']
```

```python
# splitting the data into train and test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = heart_disease_df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
model_all = RandomForestClassifier(random_state=42)
model_all.fit(x_train, y_train)

pred_all = model_all.predict(x_test)

acc_all = accuracy_score(y_test, pred_all)
print("Accuracy with all features:", acc_all)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
pred = model_all.predict(x_test)
cm = confusion_matrix(y_test, pred)
ConfusionMatrixDisplay(cm).plot()
plt.title("Confusion Matrix")
plt.show()
```

# Feature Selection

```python

from sklearn.ensemble import RandomForestClassifier

X = heart_disease_df.drop("target", axis=1)
y = heart_disease_df["target"]

model = RandomForestClassifier()
model.fit(X, y)

importances = model.feature_importances_

for feature, score in zip(X.columns, importances):
    print(feature, score)
```

```python
X_reduced = X.drop(["fbs", "restecg"], axis=1)
```

```python
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_reduced, y, test_size=0.2, random_state=42
)

model_red = RandomForestClassifier(random_state=42)
model_red.fit(X_train_r, y_train_r)

pred_red = model_red.predict(X_test_r)

acc_red = accuracy_score(y_test_r, pred_red)
print("Accuracy after removing features:", acc_red)
```

```python
cm=confusion_matrix(y_test_r,pred_red)
ConfusionMatrixDisplay(cm).plot()
plt.title("Confusion Matrix")
```

```python
print("Accuracy with all features:", acc_all)
print("Accuracy after removing features:", acc_red)
```

```python
selected = ['sex','cp','thal','exang','ca','slope','oldpeak']

X_sel = heart_disease_df[selected]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_sel, y, test_size=0.2, random_state=42
)

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)

from sklearn.metrics import accuracy_score
pred = model.predict(X_test)

print("Accuracy Score after feature selection:", accuracy_score(y_test, pred))
```

```python
cm=confusion_matrix(y_test,pred)
ConfusionMatrixDisplay(cm).plot()
plt.title("Confusion Matrix")
```

Saving the trained model

```python
import joblib

# Save the trained model to a file
joblib_file = "heart_disease_model.pkl"
joblib.dump(model_red, joblib_file)

print(f"Model saved as {joblib_file}")
```

```python
# Load the saved model
loaded_model = joblib.load("heart_disease_model.pkl")

# Make predictions using the loaded model
input_data_as_numpy_array = np.asarray(input)
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = loaded_model.predict(input_data_reshaped)

if prediction[0] == 0:
    print("The person doesn't have any heart disease")
else:
    print("The person has heart disease")
```
